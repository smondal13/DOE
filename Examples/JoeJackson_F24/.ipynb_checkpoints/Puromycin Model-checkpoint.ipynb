```python
# Linear Transformation

import numpy as np
import pyomo.environ as pyo
import matplotlib.pyplot as plt

# Provided data
substrate_concentration = np.array([0.02, 0.06, 0.11, 0.22, 0.56, 1.10])
treated_velocity = np.array([[76, 47], [97, 107], [123, 139], [159, 152], [191, 201], [207, 200]])
untreated_velocity = np.array([[67, 51], [84, 86], [98, 115], [131, 124], [144, 158], [160, np.nan]])

# Calculate the average velocities
treated_avg_velocity = np.nanmean(treated_velocity, axis=1)
untreated_avg_velocity = np.nanmean(untreated_velocity, axis=1)

# Calculate the inverse of substrate concentration and velocity
inv_substrate_concentration = 1 / substrate_concentration
inv_treated_velocity = 1 / treated_avg_velocity
inv_untreated_velocity = 1 / untreated_avg_velocity

# Function to create and solve the Pyomo model for linear regression
def linreg(x_data, y_data):
    model = pyo.ConcreteModel()

    # Indices for the data points
    model.N = pyo.RangeSet(len(x_data))

    # Decision variables for the slope and intercept
    model.slope = pyo.Var(within=pyo.Reals, initialize=1.0)
    model.intercept = pyo.Var(within=pyo.Reals, initialize=1.0)

    # Objective function: Minimize the sum of squared errors
    def objective_rule(model):
        return sum((y_data[i-1] - (model.slope * x_data[i-1] + model.intercept))**2 for i in model.N)

    model.objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

    # Solve the optimization problem
    solver = pyo.SolverFactory('ipopt')
    solver.solve(model, tee=True)

    return model.slope.value, model.intercept.value

# Fit the Lineweaver-Burk plot for the treated and untreated data
slope_treated, intercept_treated = linreg(inv_substrate_concentration, inv_treated_velocity)
slope_untreated, intercept_untreated = linreg(inv_substrate_concentration, inv_untreated_velocity)

# Plot the linearized data and the best-fit lines
plt.figure(figsize=(10, 6))
plt.scatter(inv_substrate_concentration, inv_treated_velocity, color='blue', label='Treated Data')
plt.scatter(inv_substrate_concentration, inv_untreated_velocity, color='red', label='Untreated Data')

# Plot the best-fit lines using the Pyomo results
plt.plot(inv_substrate_concentration, slope_treated * inv_substrate_concentration + intercept_treated, color='blue', linestyle='--',
         label=f'Treated Fit (Slope={slope_treated:.6f}, Intercept={intercept_treated:.6f})')
plt.plot(inv_substrate_concentration, slope_untreated * inv_substrate_concentration + intercept_untreated, color='red', linestyle='--',
         label=f'Untreated Fit (Slope={slope_untreated:.6f}, Intercept={intercept_untreated:.6f})')

# Adding the equations to the graph
equation_treated = f"1/V = {slope_treated:.6f} * (1/[S]) + {intercept_treated:.6f}"
equation_untreated = f"1/V = {slope_untreated:.6f} * (1/[S]) + {intercept_untreated:.6f}"

plt.text(0.05, 0.01, equation_treated, color='blue', fontsize=10, verticalalignment='bottom')
plt.text(0.05, 0.005, equation_untreated, color='red', fontsize=10, verticalalignment='bottom')

plt.xlabel('1 / Substrate Concentration (1/ppm)')
plt.ylabel('1 / Velocity (min²/counts)')
plt.title('Lineweaver-Burk Plot (Linearized Michaelis-Menten using Pyomo)')
plt.legend()
plt.grid(True)
plt.show()

### Getting Thetas from slope and intercept ###
# Intercept = 1/theta1
# Slope = theta2/theta1

# Untreated
theta1 = 1/intercept_untreated
theta2 = slope_untreated*theta1
print(f"Untreated: Theta1 is {theta1}, Theta2 is {theta2}")

# Treated
theta1 = 1/intercept_treated
theta2 = slope_treated*theta1
print(f"Treated: Theta1 is {theta1}, Theta2 is {theta2}")
```

```python
# Provided data (made easier to access)
substrate_concentration = np.array([0.02, 0.06, 0.11, 0.22, 0.56, 1.10])
treated_velocity = np.array([[76, 47], [97, 107], [123, 139], [159, 152], [191, 201], [207, 200]])
untreated_velocity = np.array([[67, 51], [84, 86], [98, 115], [131, 124], [144, 158], [160, np.nan]])

# Calculate the average velocities
treated_avg_velocity = np.nanmean(treated_velocity, axis=1)
untreated_avg_velocity = np.nanmean(untreated_velocity, axis=1)

# Function to create and solve the Pyomo model
def fit_michaelis_menten_pyomo(substrate_concentration, velocity_data):
    model = pyo.ConcreteModel()

    # Indices for the data points
    model.N = pyo.RangeSet(len(substrate_concentration))

    # Decision variables
    model.Vmax = pyo.Var(within=pyo.NonNegativeReals, initialize=200)
    model.Km = pyo.Var(within=pyo.NonNegativeReals, initialize=0.1)

    # Objective function: Minimize the sum of squared errors
    def objective_rule(model):
        return sum((velocity_data[i-1] - (model.Vmax * substrate_concentration[i-1]) / (model.Km + substrate_concentration[i-1]))**2 for i in model.N)

    model.objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

    # Solve the optimization problem
    solver = pyo.SolverFactory('ipopt')
    solver.solve(model, tee=True)

    return model.Vmax.value, model.Km.value

# Fit the Michaelis-Menten model to the treated and untreated data
Vmax_treated, Km_treated = fit_michaelis_menten_pyomo(substrate_concentration, treated_avg_velocity)
Vmax_untreated, Km_untreated = fit_michaelis_menten_pyomo(substrate_concentration, untreated_avg_velocity)

# Plot the data and the fitted curves
plt.figure(figsize=(10, 6))
plt.scatter(substrate_concentration, treated_avg_velocity, color='blue', label='Treated Data')
plt.scatter(substrate_concentration, untreated_avg_velocity, color='red', label='Untreated Data')

# Plot the fitted Michaelis-Menten curves using the Pyomo results
S_fit = np.linspace(0, max(substrate_concentration), 100)
plt.plot(S_fit, (Vmax_treated * S_fit) / (Km_treated + S_fit), color='blue', linestyle='--', label=f'Treated Fit (Vmax={Vmax_treated:.6f}, Km={Km_treated:.6f})')
plt.plot(S_fit, (Vmax_untreated * S_fit) / (Km_untreated + S_fit), color='red', linestyle='--', label=f'Untreated Fit (Vmax={Vmax_untreated:.6f}, Km={Km_untreated:.6f})')

plt.xlabel('Substrate Concentration (ppm)')
plt.ylabel('Velocity (counts/min²)')
plt.title('Michaelis-Menten Model Fit using Pyomo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of theta1 and theta2 values
theta1_values = np.linspace(0.1, 100, 100)  # Avoid zero
theta2_values = np.linspace(0.1, 100, 100)  # Avoid zero

# Create a meshgrid for theta1 and theta2
Theta1, Theta2 = np.meshgrid(theta1_values, theta2_values)

# Calculate eta values based on the provided formula
eta1 = Theta1 * 1.1 / (Theta2 + 1.1)
eta2 = Theta1 * 0.56 / (Theta2 + 0.56)
eta3 = Theta1 * 0.22 / (Theta2 + 0.22)

# Create multiple views
angles = [(30, 30), (30, 90), (30, 150), (60, 60), (90, 30)]

# Plot each view in a separate figure
for i, (elev, azim) in enumerate(angles):
    plt.figure(figsize=(12, 8))
    ax = plt.axes(projection='3d')

    # Plot the expectation surface
    ax.plot_surface(eta1, eta2, eta3, cmap='viridis', edgecolor='none', alpha=0.8)

    # Set view angle
    ax.view_init(elev=elev, azim=azim)

    # Adding labels
    ax.set_xlabel('η1 (Response at 1.1 ppm)')
    ax.set_ylabel('η2 (Response at 0.56 ppm)')
    ax.set_zlabel('η3 (Response at 0.22 ppm)')
    ax.set_title(f'View Angle: Elevation={elev}, Azimuth={azim}')

    # Show the plot
    plt.show()
```

```python
import numpy as np

# Provided data
substrate_concentration = np.array([0.02, 0.06, 0.11, 0.22, 0.56, 1.10])
treated_velocity = np.array([[76, 47], [97, 107], [123, 139], [159, 152], [191, 201], [207, 200]])
untreated_velocity = np.array([[67, 51], [84, 86], [98, 115], [131, 124], [144, 158], [160, np.nan]])

# Flatten the treated velocities into a flat array (using both values)
observed_eta = treated_velocity.flatten()  # Use each treated velocity separately

# Initial guesses for theta1 and theta2
theta = np.array([205.0, 0.08])  # Initial guesses

# Define the model function
def model(theta):
    theta1, theta2 = theta
    return np.array([
        (theta1 * 1.1) / (theta2 + 1.1),
        (theta1 * 0.56) / (theta2 + 0.56),
        (theta1 * 0.22) / (theta2 + 0.22),
        (theta1 * 0.11) / (theta2 + 0.11),
        (theta1 * 0.06) / (theta2 + 0.06),
        (theta1 * 0.02) / (theta2 + 0.02)
    ])

# Define the residuals
def residuals(theta): ############# Issure is here
    model_outputs = model(theta)
    return model_outputs - observed_eta[:6]  # Compare model outputs to the first 6 treated velocities

# Define the Jacobian matrix
def jacobian(theta):
    theta1, theta2 = theta
    j = np.zeros((6, 2))  # 6 data points and 2 parameters

    # Derivatives with respect to theta1 and theta2
    j[0, 0] = 1.1 / (theta2 + 1.1)  # d(eta1)/d(theta1)
    j[0, 1] = -theta1 * 1.1 / (theta2 + 1.1)**2  # d(eta1)/d(theta2)
    j[1, 0] = 0.56 / (theta2 + 0.56)  # d(eta2)/d(theta1)
    j[1, 1] = -theta1 * 0.56 / (theta2 + 0.56)**2  # d(eta2)/d(theta2)
    j[2, 0] = 0.22 / (theta2 + 0.22)  # d(eta3)/d(theta1)
    j[2, 1] = -theta1 * 0.22 / (theta2 + 0.22)**2  # d(eta3)/d(theta2)
    j[3, 0] = 0.11 / (theta2 + 0.11)  # d(eta4)/d(theta1)
    j[3, 1] = -theta1 * 0.11 / (theta2 + 0.11)**2  # d(eta4)/d(theta2)
    j[4, 0] = 0.06 / (theta2 + 0.06)  # d(eta5)/d(theta1)
    j[4, 1] = -theta1 * 0.06 / (theta2 + 0.06)**2  # d(eta5)/d(theta2)
    j[5, 0] = 0.02 / (theta2 + 0.02)  # d(eta6)/d(theta1)
    j[5, 1] = -theta1 * 0.02 / (theta2 + 0.02)**2  # d(eta6)/d(theta2)

    return j

# Gauss-Newton method with regularization
def gauss_newton(initial_theta, max_iterations=100, tol=1e-6):
    theta = initial_theta
    for iteration in range(max_iterations):
        r = residuals(theta)
        J = jacobian(theta)

        # Print for debugging
        print(f"Iteration {iteration + 1}:")
        print(f"Current theta: {theta}")
        print(f"Residuals: {r}")
        print(f"Jacobian: \n{J}")

        try:
            # Update rule using least squares solution
            delta, _, _, _ = np.linalg.lstsq(J, r, rcond=None)
            theta += delta

            # Check convergence
            if np.linalg.norm(delta) < tol:
                break
        except np.linalg.LinAlgError:
            print("Jacobian is singular, cannot compute update.")
            break

    return theta

# Set NumPy print options to avoid scientific notation
np.set_printoptions(suppress=True, precision=6)

# Estimate theta values using Gauss-Newton
estimated_theta = gauss_newton(theta)
print("Estimated θ1 and θ2:", estimated_theta)
```
