```python
# Minimize Least Squares
import pandas as pd
import pyomo.environ as pyo
import numpy as np
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Age (years)': [1, 1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 5, 6, 6, 6, 7, 7, 7, 8, 8, 8, 9, 11, 12, 12, 12],
    'PCB Concentration (ppm)': [0.6, 1.6, 0.5, 1.2, 2.0, 1.3, 2.5, 2.2, 2.4, 1.2, 3.5, 4.1, 5.1, 5.7, 3.4, 9.7, 8.6, 4.0, 5.5, 10.5, 17.5, 13.4, 4.5, 30.4, 12.4, 13.4, 26.2, 7.4]
}

df = pd.DataFrame(data)

# Take the natural log of PCB Concentration
df['ln_PCB_Concentration'] = np.log(df['PCB Concentration (ppm)'])

# Create the Pyomo model
model = pyo.ConcreteModel()

# Declare variables for coefficients (m and b)
model.m = pyo.Var(initialize=0, bounds=(-10, 10))  # Slope
model.b = pyo.Var(initialize=0, bounds=(-10, 10))  # Intercept

# Define the number of data points
N = len(df)
model.I = pyo.RangeSet(N)

# Define the likelihood function
def likelihood_rule(model):
    return sum((df['ln_PCB_Concentration'][i-1] - (model.m * df['Age (years)'][i-1] + model.b))**2 for i in model.I)

# Define the objective function to minimize the sum of squared errors
model.objective = pyo.Objective(expr=likelihood_rule(model), sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solution = solver.solve(model)

# Display results
print(f"m (slope) = {pyo.value(model.m)}")
print(f"b (intercept) = {pyo.value(model.b)}")
print(f"Objective Value = {pyo.value(model.objective)}")

# Create a range of ages for plotting the fitted line
age_range = np.linspace(df['Age (years)'].min(), df['Age (years)'].max(), 100)

# Predict ln(PCB Concentration) using the fitted model
predicted_ln_concentration = pyo.value(model.m) * age_range + pyo.value(model.b)

# Convert predicted ln concentrations back to the original scale for plotting
predicted_concentration = np.exp(predicted_ln_concentration)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Age (years)'], df['PCB Concentration (ppm)'], color='blue', label='Observed Data', alpha=0.7)
plt.plot(age_range, predicted_concentration, color='red', label='Fitted Line (y = mx + b)', linewidth=2)
plt.title('Best Fit Linear Model for PCB Concentration in Rainbow Trout by Age', fontsize=16)
plt.xlabel('Age (years)', fontsize=14)
plt.ylabel('PCB Concentration (ppm)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Bayesian Inference
import pandas as pd
import pyomo.environ as pyo
import numpy as np
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Age (years)': [1, 1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 5, 6, 6, 6, 7, 7, 7, 8, 8, 8, 9, 11, 12, 12, 12],
    'PCB Concentration (ppm)': [0.6, 1.6, 0.5, 1.2, 2.0, 1.3, 2.5, 2.2, 2.4, 1.2, 3.5, 4.1, 5.1, 5.7, 3.4, 9.7, 8.6, 4.0, 5.5, 10.5, 17.5, 13.4, 4.5, 30.4, 12.4, 13.4, 26.2, 7.4]
}

df = pd.DataFrame(data)

# Take the natural log of PCB Concentration
df['ln_PCB_Concentration'] = np.log(df['PCB Concentration (ppm)'])

# Create the Pyomo model
model = pyo.ConcreteModel()

# Declare variables for the slope and intercept with prior distributions (initial values)
model.m = pyo.Var(initialize=0, bounds=(-10, 10))  # Slope
model.b = pyo.Var(initialize=0, bounds=(-10, 10))  # Intercept

# Define the number of data points
N = len(df)
model.I = pyo.RangeSet(N)

# Define the likelihood function
def likelihood_rule(model):
    return sum((df['ln_PCB_Concentration'][i-1] - (model.m * df['Age (years)'][i-1] + model.b))**2 for i in model.I)

# Define the penalty for prior distributions (L2 penalty for priors)
def prior_penalty(model):
    return (model.m**2 + model.b**2)  # L2 regularization

# Define the objective function to maximize the posterior
model.objective = pyo.Objective(expr=likelihood_rule(model) + prior_penalty(model), sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solution = solver.solve(model)

# Display results
m_value = pyo.value(model.m)
b_value = pyo.value(model.b)
print(f"m (slope) = {m_value}")
print(f"b (intercept) = {b_value}")
print(f"Objective Value = {pyo.value(model.objective)}")

# Create a range of ages for plotting the fitted line
age_range = np.linspace(df['Age (years)'].min(), df['Age (years)'].max(), 100)

# Predict ln(PCB Concentration) using the fitted model
predicted_ln_concentration = m_value * age_range + b_value

# Convert predicted ln concentrations back to the original scale for plotting
predicted_concentration = np.exp(predicted_ln_concentration)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Age (years)'], df['PCB Concentration (ppm)'], color='blue', label='Observed Data', alpha=0.7)
plt.plot(age_range, predicted_concentration, color='red', label='Fitted Line (Bayesian)', linewidth=2)
plt.title('Bayesian Linear Model for PCB Concentration in Rainbow Trout by Age', fontsize=16)
plt.xlabel('Age (years)', fontsize=14)
plt.ylabel('PCB Concentration (ppm)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Cubic root
import pandas as pd
import pyomo.environ as pyo
import numpy as np
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Age (years)': [1, 1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 5, 6, 6, 6, 7, 7, 7, 8, 8, 8, 9, 11, 12, 12, 12],
    'PCB Concentration (ppm)': [0.6, 1.6, 0.5, 1.2, 2.0, 1.3, 2.5, 2.2, 2.4, 1.2, 3.5, 4.1, 5.1, 5.7, 3.4, 9.7, 8.6, 4.0, 5.5, 10.5, 17.5, 13.4, 4.5, 30.4, 12.4, 13.4, 26.2, 7.4]
}

df = pd.DataFrame(data)

# Take the natural log of PCB Concentration
df['ln_PCB_Concentration'] = np.log(df['PCB Concentration (ppm)'])

# Calculate the cubic root of age
df['CubicRoot_Age'] = df['Age (years)'] ** (1/3)

# Create the Pyomo model
model = pyo.ConcreteModel()

# Declare variables for the slope and intercept
model.m = pyo.Var(initialize=0, bounds=(-10, 10))  # Slope
model.b = pyo.Var(initialize=0, bounds=(-10, 10))  # Intercept

# Define the number of data points
N = len(df)
model.I = pyo.RangeSet(N)

# Define the likelihood function (sum of squared errors)
def likelihood_rule(model):
    return sum((df['ln_PCB_Concentration'][i-1] - (model.m * df['CubicRoot_Age'][i-1] + model.b))**2 for i in model.I)

# Define the objective function to minimize the likelihood (least squares)
model.objective = pyo.Objective(expr=likelihood_rule(model), sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solution = solver.solve(model)

# Display results
m_value = pyo.value(model.m)
b_value = pyo.value(model.b)
print(f"m (slope) = {m_value}")
print(f"b (intercept) = {b_value}")
print(f"Objective Value = {pyo.value(model.objective)}")

# Create a range of cubic roots of age for plotting the fitted line
cubic_root_age_range = np.linspace(df['CubicRoot_Age'].min(), df['CubicRoot_Age'].max(), 100)

# Predict ln(PCB Concentration) using the fitted model
predicted_ln_concentration = m_value * cubic_root_age_range + b_value

# Convert predicted ln concentrations back to the original scale for plotting
predicted_concentration = np.exp(predicted_ln_concentration)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['CubicRoot_Age'], df['PCB Concentration (ppm)'], color='blue', label='Observed Data', alpha=0.7)
plt.plot(cubic_root_age_range, predicted_concentration, color='red', label='Fitted Line (Least Squares)', linewidth=2)
plt.title('Least Squares Model for ln(PCB Concentration) vs. (Age)^(1/3)', fontsize=16)
plt.xlabel('Cubic Root of Age (years)', fontsize=14)
plt.ylabel('PCB Concentration (ppm)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Cubic root with confidence intervals (unsure on this: PCB 2)
import pandas as pd
import pyomo.environ as pyo
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t

# Create the DataFrame
data = {
    'Age (years)': [1, 1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4, 5, 6, 6, 6, 7, 7, 7, 8, 8, 8, 9, 11, 12, 12, 12],
    'PCB Concentration (ppm)': [0.6, 1.6, 0.5, 1.2, 2.0, 1.3, 2.5, 2.2, 2.4, 1.2, 3.5, 4.1, 5.1, 5.7, 3.4, 9.7, 8.6, 4.0, 5.5, 10.5, 17.5, 13.4, 4.5, 30.4, 12.4, 13.4, 26.2, 7.4]
}

df = pd.DataFrame(data)

# Take the natural log of PCB Concentration
df['ln_PCB_Concentration'] = np.log(df['PCB Concentration (ppm)'])

# Compute cubic root of age
df['Cubic_Root_Age'] = df['Age (years)'] ** (1/3)

# Create a Pyomo model
model = pyo.ConcreteModel()

# Declare variables for coefficients
model.a = pyo.Var(initialize=0, bounds=(-10, 10))  # slope
model.b = pyo.Var(initialize=0, bounds=(-10, 10))  # intercept

# Define the number of data points
N = len(df)
model.I = pyo.RangeSet(N)

# Define the objective function (Least Squares)
def objective_rule(model):
    return sum((df['ln_PCB_Concentration'][i-1] - (model.a * df['Cubic_Root_Age'][i-1] + model.b))**2 for i in model.I)

model.objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solution = solver.solve(model)

# Display results
a_value = pyo.value(model.a)
b_value = pyo.value(model.b)
print(f"Slope (a) = {a_value}")
print(f"Intercept (b) = {b_value}")
print(f"Objective Value = {pyo.value(model.objective)}")

# Prepare for plotting
age_range = np.linspace(df['Age (years)'].min(), df['Age (years)'].max(), 100)
predicted_ln_pcb = a_value * (age_range ** (1/3)) + b_value

# Calculate residuals and variance using evaluated coefficients
residuals = df['ln_PCB_Concentration'] - (a_value * df['Cubic_Root_Age'] + b_value)
sigma_squared = np.var(residuals, ddof=1)  # ddof=1 for sample variance

# Calculate the prediction standard error
X = np.column_stack((np.ones(N), df['Cubic_Root_Age']))  # Design matrix
predictions = np.column_stack((np.ones(len(age_range)), age_range ** (1/3)))  # For prediction
predictions_variance = predictions @ np.linalg.inv(X.T @ X) @ predictions.T
predictions_std_error = np.sqrt(sigma_squared * (1 + predictions_variance.diagonal()))

# 95% confidence intervals
t_critical = t.ppf(1 - 0.05 / 2, df=N - 2)
upper_bound = predicted_ln_pcb + t_critical * predictions_std_error
lower_bound = predicted_ln_pcb - t_critical * predictions_std_error

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(df['Cubic_Root_Age'], df['ln_PCB_Concentration'], color='blue', label='Observed Data', alpha=0.7)
plt.plot(age_range ** (1/3), predicted_ln_pcb, color='red', label='Fitted Line', linewidth=2)
plt.fill_between(age_range ** (1/3), lower_bound, upper_bound, color='orange', alpha=0.3, label='95% Confidence Interval')
plt.title('Natural Log of PCB Concentration vs. Cubic Root of Age', fontsize=16)
plt.xlabel('Cubic Root of Age (years^(1/3))', fontsize=14)
plt.ylabel('ln(PCB Concentration)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()

# Calculate confidence intervals for joint inference region
n = len(df)
beta_hat = np.array([b_value, a_value])  # Estimated coefficients

# Calculate variance-covariance matrix
XTX_inv = np.linalg.inv(X.T @ X)
cov_beta = sigma_squared * XTX_inv

# Calculate t critical value
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, df=n-2)

# Marginal confidence intervals
se = np.sqrt(np.diag(cov_beta))
ci_a = (a_value - t_critical * se[1], a_value + t_critical * se[1])
ci_b = (b_value - t_critical * se[0], b_value + t_critical * se[0])

print(f"Confidence interval for slope (a): {ci_a}")
print(f"Confidence interval for intercept (b): {ci_b}")

# Plot the joint inference region
a_range = np.linspace(ci_a[0], ci_a[1], 100)
b_range = np.linspace(ci_b[0], ci_b[1], 100)
A, B = np.meshgrid(a_range, b_range)
Z = (A - a_value) ** 2 / cov_beta[1, 1] + (B - b_value) ** 2 / cov_beta[0, 0] <= t_critical ** 2

plt.figure(figsize=(10, 6))
plt.contourf(A, B, Z, levels=[0, 1], colors=['red', 'blue'], alpha=0.3)
plt.scatter(a_value, b_value, color='black', label='Estimated Parameters', s=100)
plt.axvline(ci_a[0], color='green', linestyle='--', label='CI for a')
plt.axvline(ci_a[1], color='green', linestyle='--')
plt.axhline(ci_b[0], color='orange', linestyle='--', label='CI for b')
plt.axhline(ci_b[1], color='orange', linestyle='--')
plt.title('Joint Inference Region for Parameters', fontsize=16)
plt.xlabel('Slope (a)', fontsize=14)
plt.ylabel('Intercept (b)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyomo.environ as pyo

# Given data points
lage = np.array([1.26, 1.82, 2.22])
ln_PCB = np.array([0.92, 2.15, 2.52])

# Create a DataFrame for convenience
data = pd.DataFrame({
    'lage': lage,
    'ln_PCB': ln_PCB
})

# Create the Pyomo model
model = pyo.ConcreteModel()

# Number of data points
N = len(data)

# Declare variables for slope (a) and intercept (b)
model.a = pyo.Var(initialize=0)
model.b = pyo.Var(initialize=0)

# Define the objective function (sum of squared residuals)
def objective_rule(model):
    return sum((data['ln_PCB'][i] - (model.a * data['lage'][i] + model.b)) ** 2 for i in range(N))

model.objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solver.solve(model)

# Extract the estimated parameters
slope = pyo.value(model.a)
intercept = pyo.value(model.b)

# Display the estimated parameters
print(f"Slope (a) = {slope}")
print(f"Intercept (b) = {intercept}")

# Prepare for contour plot
a_range = np.linspace(-5, 5, 100)  # Range for slope
b_range = np.linspace(-5, 5, 100)  # Range for intercept
A, B = np.meshgrid(a_range, b_range)

# Calculate the log-likelihood for each pair (a, b)
Z = np.zeros_like(A)
sigma_squared = np.var(ln_PCB - (intercept + slope * lage))  # Variance from residuals

for i in range(len(a_range)):
    for j in range(len(b_range)):
        # Calculate the log-likelihood for the current (a, b)
        log_likelihood = -0.5 * np.sum((ln_PCB - (A[i, j] + B[i, j] * lage)) ** 2 / sigma_squared)
        Z[i, j] = log_likelihood

# Plotting the log-likelihood surface
plt.figure(figsize=(12, 8))
contour = plt.contourf(A, B, Z, levels=50, cmap='viridis')
plt.colorbar(label='Log-Likelihood')
plt.scatter(slope, intercept, color='red', label='Estimated Parameters', s=100)
plt.title('Expectation Surface for Log-Likelihood', fontsize=16)
plt.xlabel('Slope (a)', fontsize=14)
plt.ylabel('Intercept (b)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Attempt on QR Decomp
import pandas as pd
import pyomo.environ as pyo
import numpy as np
import matplotlib.pyplot as plt

# Given data points
lage = np.array([1.26, 1.82, 2.22])
ln_PCB = np.array([0.92, 2.15, 2.52])

# Create a DataFrame for convenience
data = pd.DataFrame({
    'lage': lage,
    'ln_PCB': ln_PCB
})

# Create the Pyomo model
model = pyo.ConcreteModel()

# Number of data points
N = len(data)

# Declare variables for the slope and intercept
model.a = pyo.Var(initialize=0)  # Slope
model.b = pyo.Var(initialize=0)  # Intercept

# Design matrix (X) with a column of ones for the intercept
X = np.vstack([np.ones(N), lage]).T  # Shape: (N, 2)

# Define the response vector (y)
y = ln_PCB

# QR decomposition using Pyomo
# Create the Q and R matrices as variables
model.Q = pyo.Var(range(N), range(2), within=pyo.Reals)  # Q matrix
model.R = pyo.Var(range(2), range(2), within=pyo.Reals)  # R matrix

# Define constraints for orthonormal columns of Q
def orthonormality_constraint(model, i):
    return sum(model.Q[i, j] * model.Q[i, j] for j in range(2)) == 1

model.orthonormality = pyo.Constraint(range(N), rule=orthonormality_constraint)

# Define constraints for the QR decomposition
def qr_decomposition(model, i, j):
    if i == j:
        return sum(model.Q[k, i] * y[k] for k in range(N)) == model.R[i, j]
    else:
        return sum(model.Q[k, i] * y[k] for k in range(N)) == model.R[i, j]

model.qr_constraints = pyo.Constraint(range(2), range(2), rule=qr_decomposition)

# Objective function to minimize the sum of squared residuals
def objective_rule(model):
    return sum((ln_PCB[i] - (model.a * lage[i] + model.b))**2 for i in range(N))

model.objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

# Solve the model
solver = pyo.SolverFactory('ipopt')
solution = solver.solve(model)

model.pprint()

# Display results
print(f"Slope (a) = {pyo.value(model.a)}")
print(f"Intercept (b) = {pyo.value(model.b)}")
print(f"Objective Value (Sum of Squared Residuals) = {pyo.value(model.objective)}")

# Prepare data for plotting the fitted line
age_range = np.linspace(min(lage), max(lage), 100)
predicted_ln_PCB = pyo.value(model.a) * age_range + pyo.value(model.b)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(data['lage'], data['ln_PCB'], color='blue', label='Observed Data', alpha=0.7)
plt.plot(age_range, predicted_ln_PCB, color='red', label='Fitted Line', linewidth=2)
plt.title('Linear Fit of ln(PCB) vs Age using QR Decomposition', fontsize=16)
plt.xlabel('Age (years)', fontsize=14)
plt.ylabel('ln(PCB Concentration)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()


model.pprint()
```
